```python
def one_step_forecast(df, window):
    d = df.values
    x = []
    n = len(df)
    idx = df.index[:-window]
    for start in range(n-window):
        end = start + window
        x.append(d[start:end])
    cols = [f'x_{i}' for i in range(1, window+1)]
    x = np.array(x).reshape(n-window, -1)
    y = df.iloc[window:].values
    df_xs = pd.DataFrame(x, columns=cols, index=idx)
    df_y = pd.DataFrame(y.reshape(-1), columns=['y'], index=idx)
    return pd.concat([df_xs, df_y], axis=1).dropna()
```

```python
def split_data(df, test_split=0.15):
    n = int(len(df) * test_split)
    train, test = df[:-n], df[-n:]
    return train, test

#train, test = split_data(morts_os)
#print(f'Train: {len(train)} Test: {len(test)}')
```

```python
class Standardize:
    def __init__(self, split=0.15):
        self.split = split

    def _transform(self, df):
        return (df - self.mu)/self.sigma

    def split_data(self, df):
        n = int(len(df) * test_split)
        train, test = df[:-n], df[-n:]
        return train, test

    def fit_transform(self, train, test):
        self.mu = train.mean()
        self.sigma = train.std()
        train_s = self._transform(train)
        test_s =  self._transform(test)
        return train_s, test_s

    def transform(self, df):
        return self._transform(df)

    def inverse(self, df):
        return (df * self.sigma)+self.mu

    def inverse_y(self, df):
        return (df * self.sigma[0])+self.mu[0]

#scaler = Standardize()
#train, test = scaler.fit_transform(train, test)
#train.head()
```

```python
from sklearn.linear_model import(LinearRegression,
                                 ElasticNet,
                                 HuberRegressor)
from sklearn.ensemble import (RandomForestRegressor,
                              ExtraTreesRegressor)
from sklearn.neighbors import KNeighborsRegressor

from sktime.performance_metrics.forecasting import(MeanAbsolutePercentageError,
                                                   MeanSquaredError,
                                                   MeanAbsoluteScaledError)
```

```python


morts_reg = one_step_forecast(serie_morts_log_dif2, 10)
accs_reg = one_step_forecast(serie_accs_log_dif2, 10)

train_morts, test_morts = split_data(morts_reg, test_split=0.10)
scaler_morts = Standardize()
train_morts_s, test_morts_s = scaler_morts.fit_transform(train_morts,test_morts)

train_accs, test_accs = split_data(accs_reg, test_split=0.10)
scaler_accs = Standardize()
train_accs_s, test_accs_s = scaler_accs.fit_transform(train_accs,test_accs)

print(train_accs_s.shape)
print(test_accs_s.shape)
train_accs_s.head()
```

```python
regressors = {
    'Linear Regression': LinearRegression(fit_intercept=False),
    'Elastic Net': ElasticNet(0.5, fit_intercept=False),
    'K-NN Regressor': KNeighborsRegressor(n_neighbors=5),
    'Random Forest Regressor': RandomForestRegressor()}


mse = MeanSquaredError()
mape = MeanAbsolutePercentageError()
mase = MeanAbsoluteScaledError()

def train_model(train, test, regressor, reg_name):
    X_train, y_train  = train.drop(columns=['y']), train['y']
    X_test, y_test  = test.drop(columns=['y']), test['y']

    print(f'training {reg_name} ...')

    regressor.fit(X_train, y_train)

    yhat = regressor.predict(X_test)
    rmse_test = np.sqrt(mse(y_test, yhat))
    mape_test = mape(y_test, yhat)
    mase_test = mase(y_test, yhat, y_train=y_train)
    residuals = y_test.values - yhat

    model_metadata = {
        'Model Name': reg_name, 'Model': regressor,
        'RMSE': rmse_test, 'MAPE': mape_test, 'MASE': mase_test,
        'yhat': yhat, 'resid': residuals, 'actual': y_test.values}

    return model_metadata
```

```python
def train_different_models(train, test, regressors):
    results = []
    for reg_name, regressor in regressors.items():
        results.append(train_model(train,
                                   test,
                                   regressor,
                                   reg_name))
    return results


morts_results = train_different_models(train_morts_s, test_morts_s, regressors)
accs_results = train_different_models(train_accs_s, test_accs_s, regressors)

pd.DataFrame(morts_results).sort_values('MAPE')
```

```python
cols = ['Model Name', 'RMSE', 'MAPE', 'MASE']
morts_results = pd.DataFrame(morts_results)
morts_results[cols].sort_values('MASE')
```

```python
from statsmodels.graphics.tsaplots import plot_acf
def plot_results(cols, results, data_name):
    for row in results[cols].iterrows():
        yhat, resid, actual, name = row[1]
        plt.title(f'{data_name} - {name}')
        plt.plot(actual, 'k--', alpha=0.5)
        plt.plot(yhat, 'k')
        plt.legend(['actual', 'forecast'])
        plot_acf(resid, zero=False,
                 title=f'{data_name} - Autocorrelation')
        plt.show()

cols = ['yhat', 'resid', 'actual', 'Model Name']
plot_results(cols, morts_results, 'Morts')
```
